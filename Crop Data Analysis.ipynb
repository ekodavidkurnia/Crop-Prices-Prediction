```python
import pandas as pd
data = pd.read_csv('dataset pangan.csv')

data.tail()
```

## EDA

```python
data['Tanggal'] = pd.to_datetime(data['Tanggal'], infer_datetime_format=True)
```

```python
data = data.set_index('Tanggal')
```

```python
dt = pd.date_range('2017-07-31', '2021-04-30')
idx = pd.DatetimeIndex(dt)
data = data.reindex(idx)
data.head(10)
```

```python
data.tail(10)
```

```python
data.info()
```

```python
data = data.apply(pd.to_numeric)
```

```python
data.iloc[304:307,:]
```

```python
import numpy as np
data.replace(['-'], np.nan, inplace=True)
```

```python
data.iloc[304:307,:]
```

```python
data = data.apply(pd.to_numeric)
```

## Beras

```python
import plotly.express as px

fig = px.line(data, x = data.index, y = 'Beras', title = 'Harga beras dengan slider')
fig.update_xaxes(rangeslider_visible=True,
                rangeselector=dict(
                    buttons=list([
                        dict(count=1, label='1y', step = 'year', stepmode='todate'),
                        dict(count=6, label='2m', step = 'month', stepmode='backward'),
                        dict(count=30, label='30d', step = 'day', stepmode='backward'),
                        dict(step='all')
                ])))
fig.show()
```
